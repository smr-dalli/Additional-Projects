```
# Start writing code here...
!pip install bs4
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup

cricbuzz_url = requests.get("https://www.cricbuzz.com/cricket-stats/icc-rankings/men/batting")
cricbuzz_html = soup(cricbuzz_url.content,"html.parser")
player_rank = cricbuzz_html.findAll("div",{"class":"cb-col cb-col-16 cb-rank-tbl cb-font-16"})
player_name = cricbuzz_html.findAll("a",{"class":"text-hvr-underline text-bold cb-font-16"})
player_country = cricbuzz_html.findAll("div",{"class":"cb-font-12 text-gray"})
player_points = cricbuzz_html.findAll("div",{"class":"cb-col cb-col-17 cb-rank-tbl pull-right"})

test_rankings = {}
odi_rankings = {}
t20_rankings = {}



player_rankings_test = []
player_rankings_odi = []
player_rankings_t20 = []

player_name_test = []
player_name_odi= []
player_name_t20 = []


player_country_test= []
player_country_odi= []
player_country_t20= []

player_points_test = []
player_points_odi = []
player_points_t20 = []

for players in player_rank[:99]:
    player_rankings_test.append(players.text)
for players in player_rank[99:199]:
    player_rankings_odi.append(players.text)
for players in player_rank[199:299]:
    player_rankings_t20.append(players.text)

for players in player_name[:99]:
    player_name_test.append(players.text)
for players in player_name[99:199]:
    player_name_odi.append(players.text)
for players in player_name[199:299]:
    player_name_t20.append(players.text)

for players in player_country[:99]:
    player_country_test.append(players.text)
for players in player_country[99:199]:
    player_country_odi.append(players.text)
for players in player_country[199:299]:
    player_country_t20.append(players.text)

for players in player_points[:99]:
    player_points_test.append(players.text)
for players in player_points[99:199]:
    player_points_odi.append(players.text)
for players in player_points[199:299]:
    player_points_t20.append(players.text)

test_rankings['rank'] = player_rankings_test
test_rankings['name'] = player_name_test
test_rankings['country'] = player_country_test
test_rankings['points'] = player_points_test

odi_rankings['rank'] = player_rankings_odi
odi_rankings['name'] = player_name_odi
odi_rankings['country'] = player_country_odi
odi_rankings['points'] = player_points_odi

t20_rankings['rank'] = player_rankings_t20
t20_rankings['name'] = player_name_t20
t20_rankings['country'] = player_country_t20
t20_rankings['points'] = player_points_t20

Test = pd.DataFrame(test_rankings)
print(Test)
odi = pd.DataFrame(odi_rankings)
print(odi)
t20 = pd.DataFrame(t20_rankings)
t20




```
